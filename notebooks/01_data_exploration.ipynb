```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Plot styling
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("üìä Sentiment Analysis Project - Data Exploration")
print("=" * 50)
```

```python
# Load the dataset
df = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\Sentiment_analysis_project_\data\raw\IMDB Dataset.csv')

# Basic information
print(f"üìà Dataset Shape: {df.shape}")
print(f"üìã Columns: {df.columns.tolist()}")
print(f"üíæ Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print("\n" + "="*30)

# Preview first few rows
df.head()
```

```python
print("‚ùì Missing Values:")
print(df.isnull().sum())
```

```python
sentiment_counts = df['sentiment'].value_counts()

print("üìä Sentiment Distribution:")
print(sentiment_counts)

# Plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bar
sentiment_counts.plot(kind='bar', ax=ax1, color=['skyblue', 'lightcoral'])
ax1.set_title('Sentiment Distribution - Bar Chart')
ax1.set_xlabel('Sentiment')
ax1.set_ylabel('Count')

# Pie
ax2.pie(sentiment_counts.values, labels=sentiment_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
ax2.set_title('Sentiment Distribution - Pie Chart')

plt.tight_layout()
plt.show()
```

```python
# Add new columns: review character length and word count
df['review_length'] = df['review'].str.len()
df['word_count'] = df['review'].str.split().str.len()

# Show statistics grouped by sentiment
df.groupby('sentiment')[['review_length', 'word_count']].describe()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot: Review length by sentiment
df.boxplot(column='review_length', by='sentiment', ax=ax1)
ax1.set_title('Review Length by Sentiment')
ax1.set_xlabel('Sentiment')
ax1.set_ylabel('Character Count')

# Boxplot: Word count by sentiment
df.boxplot(column='word_count', by='sentiment', ax=ax2)
ax2.set_title('Word Count by Sentiment')
ax2.set_xlabel('Sentiment')
ax2.set_ylabel('Word Count')

# Tidy up layout
plt.tight_layout()
plt.suptitle('')
plt.show()
```
